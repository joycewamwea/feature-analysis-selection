#### analyze columns against each other

```python
sns.pairplot( data=df, vars=('Siblings/Spouses Aboard','Pclass','Age','Parents/Children Aboard','Fare'), hue='Survived' )
```

#### checking nulls

```python
sns.heatmap(df.isnull(),cbar=False,yticklabels=False,cmap = 'viridis')
```

#### finding correlated variables

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(),cmap='Blues',annot=False) 
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(),cmap='Blues',annot=True) 
```

#### check skewness for numerical values

```python
df=df[['Pclass','Age','Fare']]
l = df.columns.values
number_of_columns=len(df.columns)
number_of_rows=len(l)-1/number_of_columns
plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.distplot(df[l[i]],kde=True) 
```
